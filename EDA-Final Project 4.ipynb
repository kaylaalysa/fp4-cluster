#### Numeric VS Target

```python
plt.figure(figsize=(7, 6))
sns.distplot(df.BALANCE[df.TENURE == 12], color="g", label="Tunure = 12")
sns.distplot(df.BALANCE[df.TENURE == 11], color="b", label="Tunure = 11")
sns.distplot(df.BALANCE[df.TENURE == 10], color="r", label="Tunure = 10")
plt.legend();
```

```python
cat_var = ["CASH_ADVANCE", "PURCHASES_FREQUENCY", "ONEOFF_PURCHASES_FREQUENCY", "PURCHASES_INSTALLMENTS_FREQUENCY", "CASH_ADVANCE_FREQUENCY"]


fig, axes = plt.subplots(2, 3, figsize=(15, 10))
for cat, ax in zip(cat_var, axes.flatten()):
    sns.countplot(cat, data=df, hue="TENURE", ax=ax)
```
